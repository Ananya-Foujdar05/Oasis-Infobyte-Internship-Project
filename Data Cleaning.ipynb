# Data Cleaning on New York Airbnb Data

```python
import pandas as pd
import numpy as np
```

```python
# loading the data
df = pd.read_csv("NY Airbnb.csv")
df.head()
```

```python
df.shape
```

## Duplicate Removal

```python
# checking duplicate
df.duplicated().sum()
```

##### There is no Duplicate value

```python
df.info()
```

### checking missing values

```python
df.isnull().sum()
```

## Handling the missing values

```python
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
df["last_review"] = df["last_review"].fillna("not reviewed")
df["host_name"] = df["host_name"].fillna("not available")
df["name"] = df["host_id"].fillna("not available")
df
```

```python
df.isnull().sum()
```

## Consistent formatting and units across the dataset for accurate analysis

```python
df["id"] = df["id"].astype(str)
df["host_id"] = df["host_id"].astype(str)
df["last_review"] = pd.to_datetime(df["last_review"], format = "%d-%m-%Y")
df.info()
```

# Outlier detection

```python
# Calculating mean and standard deviation
mean = df["price"].mean()
std_dev = df["price"].std()

# Calculating Z-score for each data point
df['z_score'] = (df["price"] - mean) / std_dev

# Identify outliers (threshold = 3)
outliers = df[(df["z_score"] > 3) | (df["z_score"] < -3)]

print(outliers)
```

```python
# Filtering out outliers
data_filtered = df[(df['z_score'] <= 3) & (df["z_score"] >= -3)]

print(data_filtered)
```

```python
# Dropping the z_score column 
data_filtered = data_filtered.drop(columns=["z_score"])
print(data_filtered)
```

```python
# saving the clean data
data_filtered.to_csv("clean_data.csv")
```
